1. matplotlib.image

2. Pillow

3. OpenCV (cv2)

```python
# importando o modulo de imagem do matplotlib library

import matplotlib.image as mpimg
import matplotlib.pyplot as plt
```

```python
# carregando a imagem com matplotlib.image modulo

img = mpimg.imread('/content/puppy-care-guide-for-new-parents.jpg.1')
```

```python
type(img)
```

```python
print(img.shape)
```

```python
print(img)
```

```python
# display da imagem do numpy array

img_plot = plt.imshow(img)
plt.show()
```

Redimensionando a imagem ussando Pillow library

```python
from PIL import Image
```

```python
img = Image.open('/content/puppy-care-guide-for-new-parents.jpg.1')
img_resized = img.resize((200, 200))
```

```python
img_resized.save('dog_image_resized.jpg')
```

```python
# display da imagem do numpy array

img_res = mpimg.imread('/content/dog_image_resized.jpg')
img_res_plot = plt.imshow(img_res)
plt.show()
```

```python
print(img_res.shape)
```

**Convertendo RGB imagens para Grayscale usando OpenCV**

```python
import cv2
```

```python
img = cv2.imread('/content/puppy-care-guide-for-new-parents.jpg.1')
```

```python
type(img)
```

```python
img.shape
```

```python
grayscale_image = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
```

```python
type(grayscale_image)
```

```python
grayscale_image.shape
```

cv2.imshow() will display the image. But this will not be allowed in Google Colab.

**from google.colab.patches import cv2_imshow**

```python
from google.colab.patches import cv2_imshow
```

```python
# displaying the image
cv2_imshow(grayscale_image)
```

```python
# salvando a grayscale imagem
cv2.imwrite('dog_grayscale_image.jpg', grayscale_image)
```
